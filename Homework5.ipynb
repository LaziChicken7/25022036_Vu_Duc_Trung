```python
#W5A3

def prime(a):
    for i in range(2, int(a ** 0.5) + 1):
        if a % i == 0:
            return 0
    
    return 1

def main():
    try:
        a = int(input("Nhap so nguyen a:"))

    except ValueError:
        print("Dau vao khong hop le, vui long nhap so nguyen")
        return

    if prime(a):
        print(a, "la mot so nguyen to")
    else:
        print(a, "khong la mot so nguyen to")
    return

main()
```

```python
#W5A4

def perf_num(a):
    res = 0
    for i in range(1, int(a ** 0.5) + 1):
        if a % i == 0:
            res += i
            if a // i != i:
                res += a // i

    return (res - a == a)

def main():
    try:
        a = int(input("Nhap so nguyen a"))

    except ValueError:
        print("Dau vao khong hop le, vui long nhap so nguyen")
        return
    
    if perf_num(a):
        print(a, "la mot so hoan hao")
    
    else:
        print(a, "khong la mot so hoan hao")
    
    return

main()
```

```python
#W5A6

def main():
    try:
        n = int(input("Nhap so nguyen n"))
    
    except ValueError:
        print("Dau vao khong hop le, vui long nhap so nguyen")
        return

    res = 1
    for i in range(1, n + 1):
        res *= i
    
    print(f'{n}! = {res}')

main()
```

```python
#W5A7

def solve(a, b):
    print(f'{a} + {b} = {a + b}')
    print(f'{a} - {b} = {a - b}')
    print(f'{a} / {b} = {a / b:.2f}')
    print(f'{a} * {b} = {a * b}')

def main():
    a, b = map(int, input("Nhap hai so nguyen a va b").split())
    solve(a, b)

main()
```

```python
#W5A8

def Hamming(a, b):
    d1 = 0; d2 = 0
    x = a; y = b
    while (x > 0):
        d1 += 1
        x //= 2
    while (y > 0):
        d2 += 1
        y //= 2
    
    res = 0
    for i in range(0, max(d1, d2) + 1):
        if (a >> i & 1) != (b >> i & 1): res += 1
    
    print(f'Khoang cach Hamming giua hai so {a} va {b} la {res}')

def main():
    a, b = map(int, input("Nhap hai so nguyen x va y").split())
    Hamming(a, b)

main()
```

```python
#W5A9

def sum_digit(n):
    res = 0
    while n > 0:
        res += n % 10
        n //= 10
    return res

def main():
    n = int(input("Nhap so nguyen duong n"))
    print(f'Tong chu so cua {n} la {sum_digit(n)}')

main()
```

```python
#W5A10

def solve(a, b):
    mp = [0] * 130
    for i in range(0, len(a)):
        if mp[ord(a[i])] != 0 and mp[ord(a[i])] != ord(b[i]):
            return False
        mp[ord(a[i])] = ord(b[i])
    
    return True
        


def main():
    a, b = map(str, input("Nhap hai xau a va b").split())
    if (len(a) != len(b)):
        print("False")
        return
    print(solve(a, b))

main()
    
```
